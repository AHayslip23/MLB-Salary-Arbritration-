```python
cluster1 = salary2[["Age", "Service Time"]] 
```

```python
cluster1
```

```python
X = cluster1.iloc[:, [0, 1]].values 
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title("The Elbow Method")
plt.xlabel("The Number of Clusters")
plt.ylabel("WCSS")
plt.show() 
```

```python
#use 4 clusters 
```

```python
kmeans = KMeans(n_clusters=4, init="k-means++", random_state=42)
y_kmeans = kmeans.fit_predict(X)  
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=100, c="red", label="Cluster 1")
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=100, c="blue", label="Cluster 2")
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=100, c="green", label="Cluster 3")
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=100, c="purple", label="Cluster 4")
plt.title("Clusters of Age and Service Time")
plt.xlabel("Age")
plt.ylabel("Service Time")
plt.legend()
plt.show() 
```

```python
#cluster 1 are players between 28 and 30 years old and between 2 to 4 years of service time 
#cluster 2 are young people between 24 and 26 between 2 to 3 years of service time 
#cluster 3 are between 30 and 32 years old and between 4 to 5 years of service time 
#cluster 4 are between 26 and 28 years old and between 4 to 5 years of service time 
```
